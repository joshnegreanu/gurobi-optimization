```python
import numpy as np
import gurobipy as grb
```

```python
m = grb.Model("test_model")

# Define model data using numpy arrays
A = np.array([[1, 1, 0],
              [0, 1, 1]])
c = np.array([1, 2, 3])
b = np.array([1, 1])

# Add decision variables (continuous by default, lower bound 0)
x = m.addMVar(3, name="x")

# Set the objective function (maximize c @ x)
m.setObjective(c @ x, grb.GRB.MAXIMIZE)

# Add constraints (A @ x <= b)
m.addConstr(A @ x <= b, name="constraints")

# Optimize the model
m.optimize()
```

```python
# Print the results
if m.status == grb.GRB.OPTIMAL:
    print("\nOptimization successful:")
    print(f"Optimal objective value (max profit): {m.objVal}")
    print(f"Optimal variable values (x): {x.x}") # Retrieve solution as numpy array
elif m.status == grb.GRB.INFEASIBLE:
    print("Model is infeasible")
```
